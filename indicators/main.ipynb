```python
import pandas as pd
import numpy as np
import pandas_ta as ta
```

```python
D = pd.read_parquet('/workspaces/FYP24061/BTCUSDT.parquet.gzip', engine='pyarrow')
D.head()
```

## Cleaning

```python
D = D.set_index('Open Time')
D.index.name = 'Date'
D.drop(['Close Time', 'Ignore'], axis=1, inplace=True)
D
```

## Volume

```python
D['OBV'] = ta.volume.obv(D['Close'], D['Volume'])

D
```

## Trend

### Moving Average

```python
ma_value = [10,20,50,100,200]

for val in ma_value:
    D[f'MA_{val}'] = D['Close'].rolling(window=val).mean()
    
D
```

### MACD

```python
# 12, 26 EMA
# 9 EMA of MACD

EMA_12 = pd.Series(D['Close'].ewm(span=12, adjust=False).mean())
EMA_26 = pd.Series(D['Close'].ewm(span=26, adjust=False).mean())
D['MACD_12_26'] = EMA_12 - EMA_26

D['Signal_Line_9'] = D['MACD_12_26'].ewm(span=9, adjust=False).mean()
D['MACD_Histogram'] = D['MACD_12_26'] - D['Signal_Line_9']
D
```

## Strength and Momentum

### ADX

```python
D['ADX'] = ta.adx(high=D['High'], low=D['Low'], close=D['Close'])['ADX_14']

D['+DI'] = ta.adx(high=D['High'], low=D['Low'], close=D['Close'])['DMP_14']
D['-DI'] = ta.adx(high=D['High'], low=D['Low'], close=D['Close'])['DMN_14']

D['ADX_Strong'] = (D['ADX'] > 25).astype(int)

D
```

### RSI

```python
D['RSI_7'] = ta.rsi(D['Close'], length=7)
D['RSI_14'] = ta.rsi(D['Close'], length=14)
D['RSI_21'] = ta.rsi(D['Close'], length=21)

D
```

## Volatility

### Average True Range (ATR)

```python
D['ATR_7'] = ta.atr(high=D['High'], low=D['Low'], close=D['Close'], length=7)
D['ATR_14'] = ta.atr(high=D['High'], low=D['Low'], close=D['Close'], length=14)
D['ATR_21'] = ta.atr(high=D['High'], low=D['Low'], close=D['Close'], length=21)

D
```

### Bollinger Band

```python
bollinger_10 = ta.bbands(close=D['Close'], length=10, std=2)
D['BB_10_upper'] = bollinger_10['BBU_10_2.0']  
D['BB_10_lower'] = bollinger_10['BBL_10_2.0']

bollinger_20 = ta.bbands(close=D['Close'], length=20, std=2)
D['BB_20_upper'] = bollinger_20['BBU_20_2.0']  
D['BB_20_lower'] = bollinger_20['BBL_20_2.0']

D
```
